```r
ggplot(df, aes(BloodPressure, SkinThickness )) +
  geom_point(col="steelblue", size=4)+
  geom_smooth(method="lm",col="firebrick")+
  facet_wrap(~Outcome)
```

```r
ggplot(df, aes(Age, BMI, col=Outcome)) +
  geom_point(col="gray", size=4)+
  geom_line(size=1)+
  coord_cartesian(xlim=c(20.0, 50), ylim=c(0, 70))
```

```r
gg <- ggplot(df, aes(x=BloodPressure, y=Glucose)) + 
  geom_point(aes(col=Outcome), size=3) +  # Set color to vary based on Outcome categories.
  geom_smooth(method="lm", col="firebrick", size=2) + 
  coord_cartesian() + 
  labs(title="level of Glucose & BP of Dibitic and Non Dibitic ", subtitle="From Diabetes Dataset", y="Glucose", x="BloodPressure", caption="Diabetes Demographics")

gg + scale_colour_brewer(palette = "Set1")
```
